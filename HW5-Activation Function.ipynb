```python
import matplotlib.pylab as plt
import numpy as np
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

def D_sigmoid(x):
    return sigmoid(x) * (1 - sigmoid(x))
```

# Sigmoid

```python
x = np.linspace(-10,10,100)
y = sigmoid(x)
plt.plot(x,y)
plt.show()
```

# Derivative Sigmoid 

```python
x = np.linspace(-10,10,100)
y = D_sigmoid(x)
plt.plot(x,y)
plt.show()
```

```python
def tanh(x):
    return (np.exp(x/2)-np.exp(-x/2))/(np.exp(x/2)+np.exp(-x/2))

def D_tanh(x):
    return 1-tanh(x)**2
```

# tanh

```python
x = np.linspace(-10,10,100)
X = x.tolist()
x = np.reshape(X,(1,100))
#print(x)
y = []
for dig in x[0]:
    #print(dig)
    y.append(tanh(dig))
plt.plot(x[0],y)
plt.show()
```

# Derivative tanh

```python
x = np.linspace(-10,10,100)
X = x.tolist()
x = np.reshape(X,(1,100))
#print(x)
y = []
for dig in x[0]:
    #print(dig)
    y.append(D_tanh(dig))
plt.plot(x[0],y)
plt.show()
```

```python
def ReLU(x):
    return max(0,x)

def D_ReLU(x):
    return 1 if x >= 0 else 0
```

# ReLU

```python
x = np.linspace(-10,10,100)
X = x.tolist()
x = np.reshape(X,(1,100))
#print(x)
y = []
for dig in x[0]:
    #print(dig)
    y.append(ReLU(dig))
plt.plot(x[0],y)
plt.show()
```

# Derivative ReLU

```python
x = np.linspace(-10,10,100)
X = x.tolist()
x = np.reshape(X,(1,100))
#print(x)
y = []
for dig in x[0]:
    #print(dig)
    y.append(D_ReLU(dig))
plt.plot(x[0],y)
plt.show()
```

```python
def ELU(x,a):
    return x if x > 0 else a*(np.exp(x)-1)

def D_ELU(x,a):
    return 1 if x >= 0 else ELU(x,a) + a
```

# ELU

```python
x = np.linspace(-10,10,100)
X = x.tolist()
x = np.reshape(X,(1,100))
#print(x)
y = []
for dig in x[0]:
    #print(dig)
    y.append(ELU(dig,0.5))
plt.plot(x[0],y)
plt.show()
```

# Derivative ELU

```python
x = np.linspace(-10,10,100)
X = x.tolist()
x = np.reshape(X,(1,100))
#print(x)
y = []
for dig in x[0]:
    #print(dig)
    y.append(D_ELU(dig,0.5))
plt.plot(x[0],y)
plt.show()
```

```python
def Leaky_ReLU(x,a):
    return x if x > 0 else a*x

def D_Leaky_ReLU(x,a):
    return 1 if x >= 0 else a
```

# Leaky ReLU

```python
x = np.linspace(-10,10,100)
X = x.tolist()
x = np.reshape(X,(1,100))
#print(x)
y = []
for dig in x[0]:
    #print(dig)
    y.append(Leaky_ReLU(dig,0.5))
plt.plot(x[0],y)
plt.show()
```

# Derivative Leaky ReLU

```python
x = np.linspace(-10,10,100)
X = x.tolist()
x = np.reshape(X,(1,100))
#print(x)
y = []
for dig in x[0]:
    #print(dig)
    y.append(D_Leaky_ReLU(dig,0.5))
plt.plot(x[0],y)
plt.show()
```
